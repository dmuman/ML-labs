# 2. Перейменування стовпців

Задля зручності, стовпці варто перейменувати. Зробити це можна за допомогою методу `rename`, надавши йому словник з новими назвами.

Спершу переглянемо назви наших колонок через атрибут `columns`:

```python
df.columns
```

А потім методом `rename` їх перейменуємо:

```python
df.rename(columns = {
    'Student_identifier_code_for_participation_in_the_survey' : 'student_id',
    'Age_of_student_at_the_time_of_the_data_collection_process' : 'student_age',
    'Grade_level_of_the_participant_in_the_educational_program' : 'grade_level',
    'Hours_spent_studying_on_a_weekly_basis' : 'weekly_hours',
    'Preferred_study_method_in_lecture_or_practical_sessions' : 'pref_method',
    'Number_of_absent_days_in_the_current_semester' : 'absent_days',
    'Average_test_score_in_mathematical_subjects' : 'avg_score_math',
    'Student_participation_in_extracurricular_activities' : 'part_in_extra_act',
    'Preferred_programming_language_for_project_work' : 'proj_lang',
    'Satisfaction_level_with_academic_facilities_in_percentage' : 'satisfaction_percentage',
    'Plans_for_further_education_or_employment_after_graduation' : 'after_grad_plans'
}, inplace=True)
```

Додатковий аргумент `inplace=True` означає, що метод змінить оригінальний фрейм.

Тепер виведемо перші `5` рядків, аби впевнитись, що стовпці було перейменовано:

```python
df.head(5)
```

Зміни успішні!

# 3. Обробка пропусків у даних

Як ми знаємо, сет містить пропущені дані.

Спершу виведемо кількість пропусків в кожному зі стопців.
Для цього буде використано метод `isnull`, який перевіряє, чи поле має значення `NaN`, а потім метод `sum`, який порахує їхню кількість:

```python
num_missing_values = df.isnull().sum()
num_missing_values
```

Пропущені значення в кількісних стовпцях можна замінити на середні значення по кожному зі стовпців.

Пропуски можна заповнити методом `fillna`.

Середнє значення обраховується методом `mean`. Встановимо аргумент `numeric_only` на `True`, аби рахувало тільки чисельні значення. І зміни буде збережено в оригінальному фреймі.

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
df
```

Тепер видалимо усі рядки, які містять хоча б один пропусе за допомогою методу `dropna`. Оригінальний сет змінюємо.

```python
df.dropna(inplace=True)
df
```

Перевіримо кількість пропущених значень тепер:

```python
df.isnull().sum()
```

По нулях!

# 4. Фільтрація даних

Відфільтруємо студентів, чий вік більший за `20` років. Можна використати індексуваня та логічні вирази:

```python
df[df['student_age'] > 20]
```

Тепер студентів, які відвідують більше `20` годин занять на тиждень та ще й мають рівень задоволення більше `80` відсотків.

Для логічного ТА використовується оператор `&`:

```python
df[(df['weekly_hours'] > 20) & (df['satisfaction_percentage'] > 80)]
```

Тепер буде створено новий датафрейм, який міститиме лише студентів, які обрали мовою для проєкту `Python`, `C++` або `Java`.

```python
languages = ('Python', 'C++', 'Java')   # кортеж назв мов
mask = df['proj_lang'].isin(languages)  # використано метод isin для булеанового маскування оригінального фрейму
py_cpp_java_students = df[mask]         # новий фрейм з відсортованими студентами
py_cpp_java_students
```

# 5. Групування даних

Групування даних у Pandas дуже схоже до SQL. Виконати його можна за допомогою методу `groupby`.

Спочатку згрупуємо студентів за їхнім професійним рівнем:

```python
grouped_data = df.groupby('grade_level')
grouped_data
```

Тепер методом `mean` можна порахувати середню кількість годин, які кожна група витрачала на навчання, попередньо обравши колонку `'weekly_hours'`:

```python
grouped_data['weekly_hours'].mean()
```

За допомогою методу `apply` та наданій йому лямбда-функції, яка обраховує кількість `'Yes'` у колонці `'part_in_extra_act'`, можна порахувати кількість студентів у кожній групі, які беруть участь у позанавчальних активностях:

```python
grouped_data['part_in_extra_act'].apply(lambda x: (x == 'Yes').sum())
```
